```python
import requests
import statistics
import pandas as pd
import time
```

```python
url = 'https://accounts.spotify.com/api/token'

data = {
    'grant_type': 'client_credentials',
    'client_id': 'REDACTED',
    'client_secret': 'REDACTED',
}

response = requests.post(url, data=data)
ACCESS_TOKEN = response.json()['access_token']
```

```python
df = pd.read_csv('out.csv')
df.head()
```

```python
headers = {
    'Authorization': f'Bearer {ACCESS_TOKEN}'
}
```

```python
ids = []
```

```python
acousticness = []
danceability = []
duration_ms = []
energy = []
instrumentalness = []
key = []
liveness = []
loudness = []
mode = []
speechiness = []
tempo = []
time_signature = []
valence = []
```

```python
for track in enumerate(df['Track']):
    print(f'Processing {track[0]} out of {len(df["Track"])}: {track[1]}')
    params = {
        'q': track[1],
        'type': 'track',
        'limit': 1
    }
    
    res_search = requests.get('https://api.spotify.com/v1/search', params=params, headers=headers).json()
    try:
        ids.append(res_search['tracks']['items'][0]['id'])
    except:
        ids.append('')
    # print(track_id)
    print(f'Id length: {len(ids)}')
df['Spotify ID'] = ids
df.to_csv('out.csv', index=False)
      
    
```

# <span style="color: red"> CAUTION! </span>
## This can very easily rate-limit the API

```python
assert(len(acousticness) == 0)
assert(len(danceability) == 0)
assert(len(duration_ms) == 0)
assert(len(energy) == 0)
assert(len(instrumentalness) == 0)
assert(len(key) == 0)
assert(len(liveness) == 0)
assert(len(loudness) == 0)
assert(len(mode) == 0)
assert(len(speechiness) == 0)
assert(len(tempo) == 0)
assert(len(time_signature) == 0)
assert(len(valence) == 0)

chunked_ids = [ids[i: i+100] for i in range(0, len(ids), 100)]

idx = 0
for chunk in chunked_ids:
    print(f'Processing chunk {idx}')
    prepared_stmnt = ''
    for this_id in chunk:
        prepared_stmnt += this_id + ','
    # print(prepared_stmnt)
    
    time.sleep(1)
    features = requests.get(f'https://api.spotify.com/v1/audio-features?ids={prepared_stmnt}', headers=headers).json()
    print(features)
    for e in features['audio_features']:
        try:
            if e is None:
                print('EMPTY ID')
                acousticness.append(None)
                danceability.append(None)
                duration_ms.append(None)
                energy.append(None)
                instrumentalness.append(None)
                key.append(None)
                liveness.append(None)
                loudness.append(None)
                mode.append(None)
                speechiness.append(None)
                tempo.append(None)
                time_signature.append(None)
                valence.append(None)
                
            else:
                acousticness.append(e['acousticness'])
                danceability.append(e['danceability'])
                duration_ms.append(e['duration_ms'])
                energy.append(e['energy'])
                instrumentalness.append(e['instrumentalness'])
                key.append(e['key'])
                liveness.append(e['liveness'])
                loudness.append(e['loudness'])
                mode.append(e['mode'])
                speechiness.append(e['speechiness'])
                tempo.append(e['tempo'])
                time_signature.append(e['time_signature'])
                valence.append(e['valence'])
        except:
            print('DUMPED: ')
            print(e)
            break
    
    idx+=1
        
```

```python
df['acousticness'] = acousticness
df['danceability'] = danceability
df['duration_ms'] = duration_ms
df['energy'] = energy
df['instrumentalness'] = instrumentalness
df['key'] = key
df['liveness'] = liveness
df['loudness'] = loudness
df['mode'] = mode
df['speechiness'] = speechiness
df['tempo'] = tempo
df['time_signature'] = time_signature
df['valence'] = valence
```

```python
df.to_csv('out.csv', index=False)
```
