#### **STUDENT EXERCISE**
There are alot of columns that specify being count, likes, and reach but are being shown as Strings instead of integers. Let's change that using the [`replace()` function](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.replace.html#pandas.DataFrame.replace) and[`to_numeric()` function](https://pandas.pydata.org/docs/reference/api/pandas.to_numeric.html)

Here's an example on the `Spotify Streams` column:

```python
file_upload['Spotify Streams'].head(5)
```

Although when thinking of numbers, we use commas to separate between thousand, millions, and billions. Python doesn't like that, a comma is a character and a combination of characters and numbers will automatically be recognized as a String. We need to change the column to just see numbers first.

```python
file_upload['Spotify Streams']= file_upload['Spotify Streams'].replace(',','', regex=True)
```

```python
file_upload.info()
```

Now that python has this data available as just numbers, we need to decide what to do with NaN (Not a Number) values. 

I can replace these with 0, assuming that if 0's don't exist in the column they are purposely left blank. Or I can leave them as is.

I do this by using the operator `==` on my column looking for 0, then applying that filter to the whole dataset.

```python
file_upload['Spotify Streams'] = pd.to_numeric(file_upload['Spotify Streams'].fillna(0), downcast = 'integer', errors='coerce').fillna(0)
```

```python
file_upload[file_upload['Spotify Streams'] == 0].head(5)
```

I can see that there are columns that exist with 0, so to assume NaN columns are equal to 0 would be False. I want to keep these NaN values to not affect any mathematical calculations down the line. I'm using the `fillna()` method to fill the column to just be blank.

To review the results of our work, lets print out the first couple of columns again.

```python
file_upload.head(5)
```

```python
file_upload.dtypes
```

#### **ANSWER**

There are always multiple ways to perform the same action. We already saw how to perform it column by column above, lets see how to do it in bulk.

First, I identified all the columns and made a list in the column reference and put them in their own variable called `strings_to_nums`. This allows me to not have to repeat this long list of columns everytime I reference it in the code. We are metaphorically setting this data to the side to perform our actions on it.

```python
strings_to_nums = file_upload[['All Time Rank','Spotify Playlist Count','Spotify Playlist Reach','YouTube Views','YouTube Likes'
                               ,'TikTok Posts','TikTok Likes','TikTok Views','YouTube Playlist Reach','AirPlay Spins','SiriusXM Spins'
                               ,'Deezer Playlist Reach','Pandora Streams','Pandora Track Stations','Soundcloud Streams','Shazam Counts']].fillna(0)
```

Just like in the example, I used `replace` to remove the commas and replace them with nothing.

```python
strings_to_nums = strings_to_nums.replace(',','', regex=True)
```

Next, we want to use the `to_numeric` function again. Now that we are referencing multiple columns instead of just one, our input to the function no longer counts as a series(column). It is now considered a DataFrame(dataset).

The `to_numeric` function requires the [`apply` method](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.apply.html#pandas.DataFrame.apply) to execute. This method will apply the function to each series we wanted to convert. To specify how we want to treat the Null values, provide the answer in the `args` portion. This must be in order of how it is executed in the `to_numeric` function to work so ensure to revist documentation.

```python
strings_to_nums = strings_to_nums.apply(pd.to_numeric, args =('coerce','integer')).fillna('0')
strings_to_nums.dtypes
```

Now that we have completed the conversion, we need to move these data back into our original dataset.

```python
file_upload[['All Time Rank','Spotify Playlist Count','Spotify Playlist Reach','YouTube Views','YouTube Likes','TikTok Posts','TikTok Likes'
             ,'TikTok Views','YouTube Playlist Reach','AirPlay Spins','SiriusXM Spins','Deezer Playlist Reach','Pandora Streams','Pandora Track Stations'
             ,'Soundcloud Streams','Shazam Counts']] = strings_to_nums
```

```python
file_upload[['Spotify Popularity', 'Apple Music Playlist Count', 'Deezer Playlist Count', 'Amazon Playlist Count']] = file_upload[['Spotify Popularity', 'Apple Music Playlist Count', 'Deezer Playlist Count', 'Amazon Playlist Count']].fillna(0).astype(int)
```

### How to visualize the Top 10 most popular artist.

When learning how to use Pandas, we challenged ourselves to find the most popular artist on the list. Following up from that, let's visualize the Top 10 Most Popular artists on the list.

First, I want to breakout the data for my chart into a subset. It's easier to break larger problems into smaller ones to solve. To do this, created a *variable* called 'data_subset'. Then I referenced the data, the column, and the functions I needed.

```python
data_subset = file_upload.Artist.value_counts().head(10)
print(data_subset)
```

How did I know there was a `value_counts()` function available? Referencing the documentation. I also remembered from our earlier exercises that the `head()` function will return the top results so I specified for 10. Last, I want to see the work I did so I used `print()`.

One way to plot a chart is using the `plot` attribute on the subset of data we created. Then designating the kind. Feel free to review the [documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot) to see what else you can change this to.

We name the chart using `title` and designate our x and y axis.

Last we use `show` to view the chart.

```python
data_subset.plot(kind = 'bar', color = 'cyan')
plt.title('Top 10 Artist')
plt.xlabel('Artist')
plt.ylabel('Count')
plt.show()
```

Now we have a new insight. Within the top 10, Bad Bunny is a close runner up for the most popular artist while other artist have almost half as many tracks that made the list.

### How do the different metrics relate to each other?

A good way to determine how metrics relate to each other are correlation matricies. A positive correlation means that when one metric goes up, so does the other. A negative correlation means that when one metric goes up, another goes down. 

0 indicates no correlation and +/- 1 indicates correlation, the closer you are to those numbers tells you the strength of the relationship.

For this type of graph, we only need numeric data types so we will make a subset of data to filter for that. Using `select_dtypes` we will make a list of data types we want to include. From there, we want to actually calculate the actual values of how the metrics correlate using `corr()`

```python
numeric_df = file_upload.select_dtypes(include=['float64','int32','int64', 'int16'])
corr_matrix = numeric_df.corr()
print(corr_matrix)
```

This may feel a little in reverse, but we essentially want to set the stage for our visualizaton by defining the size of the graph and activating the color palette we want to use. We haven't done this with previous as the graph was not as complex where a smaller size makes it illegible. 

Last of all, create the heatmap using `heatmap`. Inputs are the data, opting in to the correlation values being annotated to the graph, select or color palette, and to center the color palette to show 0 as gray.

Finally, name the graph and show it!

```python
plt.figure(figsize=(20, 10))
sns.color_palette("coolwarm", as_cmap=True)

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix of Numeric Features')
plt.show()
```
